```python
import numpy as np
import pandas as pd
from pathlib import Path
```

```python
train_df = pd.read_csv(Path('../2019loans.csv'))
test_df = pd.read_csv(Path('../2020Q1loans.csv'))
```

```python
# Convert categorical data to numeric and separate target feature for training data
train_df.info()

# Remove NA values, none found




# Use get_dummies()
X_train_df = pd.get_dummies(train_df.drop(columns=["loan_status"]))
X_train_df
```

```python
X_train_df.info()
```

```python
y_train = train_df["loan_status"]
y_train
```

```python
# Convert categorical data to numeric and separate target feature for testing data
X_test_df = pd.get_dummies(test_df.drop(columns=["loan_status"]))
X_test_df
```

```python
y_test = test_df["loan_status"]
y_test
```

```python
X_test_df.info()
```

```python
X_train_df.columns
```

```python
# add missing dummy variables to testing set
for column in X_train_df.columns:
    if column not in X_test_df.columns:
        X_test_df[column] = 0
```

```python
X_test_df.info()
```

```python
# Train the Logistic Regression model on the unscaled data and print the model score
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=3)
classifier

classifier.fit(X_train_df, y_train)

classifier.score(X_test_df, y_test)
```

```python
# Train a Random Forest Classifier model and print the model score
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(random_state=3)
classifier.fit(X_train_df, y_train)
classifier.score(X_test_df, y_test)
```
